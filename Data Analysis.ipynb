# Data analysis
The purpose of this notebook is to explore all given datasets and to present some descriptive statistics.

```python
import os

import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
DATA_DIR = 'semeval-data-txt/en'
files = os.listdir(DATA_DIR)
file_paths = [os.path.join(DATA_DIR, file) for file in files]
```

## EI-reg (Emotion intensity regression)

```python
for file in [file for file in file_paths if 'EI-reg' in file]:
    print(file)
    print('-'*64)
    df = pd.read_csv(file, sep='\t')
    df['Intensity Score'] = df['Intensity Score'].apply(lambda x: None if x == 'NONE' else x)
    df['Intensity Score'] = df['Intensity Score'].astype(np.float)
    display(df[:3])
    if not 'test' in file:
        df.hist(column='Intensity Score')
        plt.show()
    print('\n\n')
```

## EI-oc (Emotion intensity ordinal classification)

```python
for file in [file for file in file_paths if 'EI-oc' in file]:
    print(file)
    print('-'*64)
    df = pd.read_csv(file, sep='\t')
    display(df[:3])
    if not 'test' in file:
        df['Intensity Class'] = df['Intensity Class'].apply(lambda x: int(x.split(':')[0]))
        df['Intensity Class'] = df['Intensity Class'].astype(np.int)
        df.hist(column='Intensity Class')
        plt.show()
    print('\n\n')
```
